```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = '/home/julisko/Seafile/My Library/DDT_Au_Paper_Ruitenbeek/Sim_preamp_significant_phaseshift.txt'
file = '/home/julisko/Seafile/My Library/test_output_GUI_SignalDetectionPhD/Filtered_Time_Trace_xAxisTime.txt'
#file = '/home/julisko/Seafile/My Library/test_output_GUI_SignalDetectionPhD/Filtered_Time_Trace.txt'
```

```python
data = pd.read_csv(file,delimiter = '\t')
print(data)
```

```python
labelsize = 40
fig, ax = plt.subplots(figsize=(15,10))

print(np.shape(data)[0])

data_np = data.transpose().to_numpy()

start_px = 190000
#end_px = np.shape(data)[0]
end_px = 209000


ax.plot(data_np[0][start_px:end_px],data_np[1][start_px:end_px],label='Filtered TT',linewidth=5,color='tab:blue')
ax.tick_params(axis='x',labelsize=labelsize)
ax.tick_params(axis='y',labelsize=labelsize,labelcolor='tab:blue',color='tab:blue')
ax.tick_params(which='major',labelsize=labelsize,length=15,width=5)
ax.tick_params(which='minor',labelsize=labelsize-20,length=7.5,width=2.5)
ax.set_xlabel('Time in s',size=labelsize)
ax.set_ylabel('Current in pA',size=labelsize,color='tab:blue')
```

```python
labelsize = 40
fig, ax = plt.subplots(figsize=(12,14))

start_px = 0
end_px = 2000

ax.plot(data['time'][start_px:end_px]*1000,data['I(I1)'][start_px:end_px]*1e12,label='Input Current',linewidth=5,color='tab:blue')
ax.tick_params(axis='x',labelsize=labelsize)
ax.tick_params(axis='y',labelsize=labelsize,labelcolor='tab:blue',color='tab:blue')
ax.tick_params(which='major',labelsize=labelsize,length=15,width=5)
ax.tick_params(which='minor',labelsize=labelsize-20,length=7.5,width=2.5)
ax.set_xlabel('Time in ms',size=labelsize)
ax.set_ylabel('Input Current in pA',size=labelsize,color='tab:blue')

ax2=ax.twinx()
ax2.tick_params(axis='y',labelsize=labelsize,labelcolor='tab:orange',color='tab:orange')
ax2.tick_params(which='major',labelsize=labelsize,length=15,width=5)
ax2.tick_params(which='minor',labelsize=labelsize-20,length=7.5,width=2.5)
ax2.set_ylabel('Output Voltage in V',size=labelsize,color='tab:orange')
ax2.plot(data['time'][start_px:end_px]*1000,data['V(n002)'][start_px:end_px],label='Output Voltage',linewidth=5,color='tab:orange')

#plt.legend(fontsize=labelsize)
```
